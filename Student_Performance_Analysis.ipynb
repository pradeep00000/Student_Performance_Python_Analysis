```python
# level of Parental_Involvement , Access_to_Resources , Motivation_Level , Family_Income , Teacher_Quality ||  
# Low = 0 , Medium = 1 , High = 2

df_num = df.copy()
df_num['Parental_Involvement'] = df['Parental_Involvement'].apply(lambda level : 0 if level == 'Low' else(1 if level == 'Medium' else 2))
df_num['Access_to_Resources'] = df['Access_to_Resources'].apply(lambda level : 0 if level == 'Low' else(1 if level == 'Medium' else 2))
df_num['Motivation_Level'] = df['Motivation_Level'].apply(lambda level : 0 if level == 'Low' else(1 if level == 'Medium' else 2))
df_num['Family_Income'] = df['Family_Income'].apply(lambda level : 0 if level == 'Low' else(1 if level == 'Medium' else 2))
df_num['Teacher_Quality'] = df['Teacher_Quality'].apply(lambda level : 0 if level == 'Low' else(1 if level == 'Medium' else 2))
df_num.head()
```

```python
# if yes = 1 , no = 0 , positive = 1 , negative = 0
# column is :  Extracurricular_Activities , Internet_Access , Peer_Influence , Learning_Disabilities 

df_num['Extracurricular_Activities'] = df_num['Extracurricular_Activities'].apply(lambda x : 1 if x == 'Yes' else  0)
df_num['Internet_Access'] = df_num['Internet_Access'].apply(lambda x : 1 if x == 'Yes' else  0)
df_num['Peer_Influence'] = df_num['Peer_Influence'].apply(lambda x : 1 if x == 'Yes' else  0)
df_num['Learning_Disabilities'] = df_num['Learning_Disabilities'].apply(lambda x : 1 if x == 'Yes' else  0)
df_num.head()
```

```python
# distance :  Near = 0 , Moderate = 1 , Far = 2 , Distance_from_Home 
# gender :  Male = 0 , Female = 1 , Gender

df_num['Distance_from_Home'] = df_num['Distance_from_Home'].apply(lambda y : 0 if y == 'Near' else(1 if y == 'Moderate' else 2))
df_num['Gender'] = df_num['Gender'].apply(lambda c : 0 if c == 'Male' else 1)
df_num.head()
```

1) What percentage of students had low family income, low teacher quality and
  low access to resources but still managed to beat the odds and
  score well in the exam. (Lets consider 75+ marks to be a good score)

```python
odds = df[(df['Family_Income'] == 'Low') &
   (df['Teacher_Quality'] == 'Low') &
   (df['Access_to_Resources'] == 'Low') &
   (df['Exam_Score'] >= 75)
  ].shape[0]
```

```python
group = df[(df['Family_Income'] == 'Low') &
   (df['Teacher_Quality'] == 'Low') &
   (df['Access_to_Resources'] == 'Low')
  ].shape[0]
percentage = odds/group * 100
percentage
```

3) Did students with more than or equal to 5 tutoring sessions benefit from it?


```python
u_tuitions = df[df['Tutoring_Sessions'] >= 5]
# lower bound tuitions = sessions < 5
l_tuitions = df[df['Tutoring_Sessions'] < 5]

print(u_tuitions['Exam_Score'].mean())
print(l_tuitions['Exam_Score'].mean())
```

4) Are extracurricular activities and physical activities affecting the student's exam scores

```python
df_act = df[(df['Extracurricular_Activities'] == 'Yes') & (df['Physical_Activity'] > 3)]
print(df_act['Exam_Score'].mean())
print(df_act['Exam_Score'].max())
```

```python
# who don't participate in activities
df_no_act = df[(df['Extracurricular_Activities'] == 'No') & (df['Physical_Activity'] == 0)]
print(df_no_act['Exam_Score'].mean())
print(df_no_act['Exam_Score'].max())
```

5) How influencial is the parent's involvement

```python
df_high = df[df['Parental_Involvement']== 'High']
df_medium = df[df['Parental_Involvement']== 'Medium']
df_low = df[df['Parental_Involvement']== 'Low']

# Now lets calculate the mean exam scores
print(df_high["Exam_Score"].mean())
print(df_medium["Exam_Score"].mean())
print(df_low["Exam_Score"].mean())
```
